## Calculus and more stuff

```python
# take a derivative
sympy.diff(sympy.sin(x) * sympy.cos(x) ** 2, x)
```

```python
# take an integral
sympy.integrate(x ** 2, x)
```

```python
# provide some bounds!
sympy.integrate(sympy.exp(-x**2), (x, -sympy.oo, +sympy.oo))
```

```python
# dirac?
t = sympy.symbols('t')
d = sympy.functions.special.delta_functions.DiracDelta(t)
d
```

```python
tau = sympy.symbols('tau')
doffs = sympy.functions.special.delta_functions.DiracDelta(t - tau)
sympy.integrate(sympy.sin(t) * doffs, (t, -sympy.oo, +sympy.oo)) # amazing!
```

```python
# limits
sympy.limit(sympy.sin(x) / x, x, 0)
```

```python
# solving equations (with rhs = 0)
sympy.solve(x**2 + 2*x+1, x)
```

```python
# solving differential equations
from sympy import diff

I = sympy.Function('i')
C, L = sympy.symbols('C L')
eq = sympy.Eq(I(t).diff(t).diff(t) + 1 / (L * C) * I(t), 0)

# regular solve => rearranging the equation
sympy.solve(eq, I(t))
```

```python
# differential solve
sympy.dsolve(eq, i(t))
```

```python
# matrix eigenvalues
sympy.Matrix([[1, 2], [2, 2]]).eigenvals()
```

## Example: Continued Fractions

```python
def list_to_frac(l):
    expr = sympy.Integer(0)
    for i in reversed(l[1:]):
        expr += i
        expr = 1/expr
    return l[0] + expr
```

```python
x, y, z = symbols('x y z')
list_to_frac([x, y, z])
```
