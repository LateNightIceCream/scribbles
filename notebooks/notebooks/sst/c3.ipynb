```python
import sympy
from sympy import symbols, I, solve, re, im
```

```python
x = symbols('x')
expr = x**2 - 4
sympy.plot(expr)
solve(expr)
```

```python
expr2 = x**2 + 4
sympy.plot(expr2)
solve(expr2)
```

```python
t = symbols('t')
expr = t**(-sympy.Rational(2, 3))
p1 = sympy.plot(expr, xlim=(0.02, 10), show=False, ylim=(0, 2000))
p2 = sympy.plot(expr**2, show=False)
p1.extend(p2);
p1.show()
```
